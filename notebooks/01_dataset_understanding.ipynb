```python
def create_basic_profile(
    dataframe: pd.DataFrame
) -> pd.DataFrame:
    profile = pd.DataFrame({
        "column_name": dataframe.columns,
        "data_type": dataframe.dtypes.astype(str).values,
        "row_count": len(dataframe),
        "null_count": dataframe.isna().sum().values,
        "null_percentage": (
            dataframe.isna().mean() * 100
        ).round(2).values,
        "unique_count": dataframe.nunique(
            dropna=False
        ).values
    })

    return profile
```

```python
import numpy as np


def profile_dataframe(
    dataframe: pd.DataFrame,
    dataset_name: str
) -> pd.DataFrame:
    profile_rows = []

    for column in dataframe.columns:
        series = dataframe[column]

        non_null_values = series.dropna()

        sample_values = (
            non_null_values
            .astype(str)
            .drop_duplicates()
            .head(3)
            .tolist()
        )

        row = {
            "dataset": dataset_name,
            "column_name": column,
            "data_type": str(series.dtype),
            "row_count": len(dataframe),
            "null_count": int(series.isna().sum()),
            "null_percentage": round(
                series.isna().mean() * 100,
                2
            ),
            "unique_count": int(
                series.nunique(dropna=False)
            ),
            "sample_values": " | ".join(sample_values)
        }

        if pd.api.types.is_numeric_dtype(series):
            row["minimum"] = (
                non_null_values.min()
                if not non_null_values.empty
                else np.nan
            )

            row["maximum"] = (
                non_null_values.max()
                if not non_null_values.empty
                else np.nan
            )

            row["mean"] = (
                round(non_null_values.mean(), 2)
                if not non_null_values.empty
                else np.nan
            )
        else:
            row["minimum"] = None
            row["maximum"] = None
            row["mean"] = None

        profile_rows.append(row)

    return pd.DataFrame(profile_rows)
```

# Initial Dataset Findings

## Dataset coverage

The project uses the Singapore Inside Airbnb dataset dated 29 June 2026.
The downloaded files include detailed listings, calendar records, detailed
reviews, summary listings, summary reviews, neighbourhood reference data,
and neighbourhood GeoJSON boundaries.

## Dataset relationships

- `listings.id` links to `calendar.listing_id`.
- `listings.id` links to `reviews.listing_id`.
- `listings.host_id` represents the host entity.
- `listings.neighbourhood_cleansed` links listings to neighbourhoods.
- Calendar and review dates can later link to a date dimension.

## Initial data-quality observations

- Some listing and host fields contain missing values.
- Price fields are stored as text and require currency cleaning.
- Percentage fields require conversion from text to numeric values.
- Date fields require standard parsing.
- Availability values require conversion from `t` and `f` to Boolean values.
- Missing review scores should not be treated as negative ratings because some
  listings have not received reviews.

## Important limitations

Calendar availability does not confirm whether a listing was booked. An
unavailable date may represent a booking, a host block, maintenance, or
another reason. Therefore, any occupancy measure created from calendar data
will be described as an estimated occupancy proxy.

```python
def create_id_set(
    dataframe: pd.DataFrame,
    column_name: str
) -> set[int]:
    return set(
        pd.to_numeric(
            dataframe[column_name],
            errors="coerce"
        )
        .dropna()
        .astype("int64")
    )


detailed_listing_ids = create_id_set(
    listings_detailed,
    "id"
)

summary_listing_ids = create_id_set(
    listings_summary,
    "id"
)

calendar_listing_ids = create_id_set(
    calendar,
    "listing_id"
)

detailed_review_listing_ids = create_id_set(
    reviews_detailed,
    "listing_id"
)

summary_review_listing_ids = create_id_set(
    reviews_summary,
    "listing_id"
)
```

```python
summary_only_ids = (
    summary_listing_ids -
    detailed_listing_ids
)

detailed_only_ids = (
    detailed_listing_ids -
    summary_listing_ids
)

print(
    "Summary-only listing IDs:",
    len(summary_only_ids)
)

print(
    "Detailed-only listing IDs:",
    len(detailed_only_ids)
)
```

# Dataset Relationship Findings

The summary listings file contains 3,247 unique listings, while the detailed
listings file contains 3,097. Therefore, 150 listings appear only in the
summary file.

The calendar contains 1,185,155 records, which is exactly 3,247 listings
multiplied by 365 calendar dates. This indicates that the calendar dataset
uses the summary listings file as its complete listing population.

A total of 150 calendar listing IDs and 103 review listing IDs are absent
from the detailed listings file. These records are not automatically treated
as invalid because they are present in the summary listings dataset.

The detailed reviews file and summary reviews file both contain 41,265 rows.
The apparent duplicates in the summary file occur because it contains only
listing ID and review date. Multiple distinct reviews may therefore appear
identical when they were submitted for the same listing on the same date.

The detailed reviews dataset will be used as the primary review source
because it contains a unique review ID and additional review attributes.
